```python
def crawler(cnt):
    global num
    global data_list
    global contents
    global title
    global user
    global date
    global content
    
    for i in range(1, cnt):
        try:
            driver.find_element_by_xpath('//*[@id="blogColl"]/div[3]/ul/li[%s]/div[2]/div/div[1]/a' %i).click()
        except:
            pass
        else:
            driver.switch_to_window(driver.window_handles[1]) 
            driver.get_window_position(driver.window_handles[1]) # 새창 넘어가기
            extract={"id": "TFRW21", 
                     "metadata": { "title": "휘뚜마뚜 웹문서 말뭉치 TFRW21", 
                                  "creator": "김자민", 
                                  "distributor": "휘뚜마뚜", 
                                  "year": "2021", 
                                  "category": "티스토리 > 음식", 
                                  "annotation_level": [ "원시" ], 
                                  "sampling": "텍스트 추출" }, 
                     "document": {"metadata": { "title": "", 
                                                 "author": "", 
                                                 "date": "" }, 
                                    "content": {"text": ""}}
                                }
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')
            no = soup.select_one('div.col-md-9')
            if no == None:
                user_no = soup.find('span', class_='jb-site-title-text')
                if soup.find('span', class_='jb-site-title-text') != None:
                    user = soup.find('span', class_='jb-site-title-text').get_text().strip()
                    try:
                        title = soup.find('div', class_='jb-content-title jb-content-title-article').find('h2').find('a').get_text().strip()
                        date = soup.find('li', class_='jb-article-information-date').find('span').get_text().strip()
                    except:
                        pass
                    contents = []
                    try:
                        contents += soup.select_one('div.jb-article').select('b')
                    except:
                        pass
                    try:
                        contents += soup.select_one('div.jb-article').select('p')
                    except:
                        pass
                elif soup.find('span', class_='author') != None:
                    user = soup.find('span', class_='author').get_text().strip()
                    try:
                        title = soup.find('div', class_='post-cover').find('h1').get_text().strip()
                    except AttributeError:
                        try:
                            title = soup.find('div', class_='hgroup').find('h1').get_text().strip()
                        except:
                            pass
                    except:
                        pass
                    try:
                        date = soup.find('span', class_='date').get_text().strip()
                    except:
                        pass
                    contents = []
                    try:
                        contents += soup.select_one('div.entry-content').select('p')
                    except:
                        pass
                elif soup.find('span', class_='writer') != None:
                    user = soup.find('span', class_='writer').get_text().strip()
                    try:
                        title = soup.find('h2', class_='title-article').get_text().strip()
                    except:
                        try:
                            title = soup.find('h1', class_='title-article').get_text().strip()
                        except:
                            pass
                    try:
                        date = soup.find('span', class_='date').get_text().strip()
                    except:
                        pass
                    contents = []
                    try:
                        contents += soup.select_one('div.article-view').select('div>p')
                    except:
                        pass
                    try:
                        contents += soup.select_one('div.article-view').select('div>h3>b')
                    except:
                        pass
                    try:
                        contents += soup.select_one('div.article-view').select('div>div>h3>span>b>span')
                    except:
                        pass
                    try:
                        contents += soup.select_one('div.article-view').select('div>p>span>b>i')
                    except:
                        pass
                    try:
                        contents += soup.select_one('div.article-view').select('div>p>span>b')
                    except:
                        pass
                    try:
                        contents += soup.select_one('div.article-view').select('div>p>span>span')
                    except:
                        pass
                else:
                    pass
                content = ''
                for i in contents:
                    if len(i.get_text()) == 1:
                        pass
                    else:
                        cont = i.get_text().strip()
                        emoji = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
                        text = emoji.sub(r' ', cont)
                        content += text.replace(u'\xa0', u'').replace(u'\n', u'').replace('<!— 서식광고 —>      (adsbygoogle = window.adsbygoogle || []).push({});','').replace('\u200b','')
                extract['id'] = "TFRW21" + str(num).zfill(8)
                extract['document']['metadata']['title'] = title
                extract['document']['metadata']['author'] = user
                extract['document']['metadata']['date'] = date
                extract['document']['content']['text'] = content
                if content == '':
                    pass
                else:
                    num += 1
                    data_list.append(extract)
                    print(extract)
                # print(user)
                # print(title)
            #     print(date)
                # print(content)
                driver.close() # 새창 닫기
                driver.switch_to.window(driver.window_handles[0]) # 기존 창으로 돌아오기
            else:
                driver.close() # 새창 닫기
                driver.switch_to.window(driver.window_handles[0]) # 기존 창으로 돌아오기

#각 게시글의 제목을 클릭합니다.
run = True
while run:
    number = int(input('크롤링할 글 개수를 입력해주세요: '))
    if number > 0:
        num = 981
        data_list = []
        for no in range(0, int(number/10)+1):
            if number%10 == 0 and int(number/10) != no:
                crawler(11)
            elif number%10 != 0:
                if int(number/10) != no:
                    crawler(11)
                else:
                    crawler(number%10+1)
            try:
                driver.find_element_by_xpath('//*[@id="pagingArea"]/span/span[3]/a').click()
            except:
                print('크롤링할 내용이 없습니다.')
                run = False
                break
    else:
        print('0이상의 수를 입력해주세요.')
        continue
        /html/body/article/article/div/article/div/div[2]/div/div[3]/ul/li[3]
```
