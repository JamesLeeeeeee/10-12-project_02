```python
import json
from collections import Counter
from wordcloud import WordCloud
```

```python
#Load in cookie
cookies = json.load(open('instagram_data_hwa_2-4gate.json', 'rb'))
# 좋아요, 해쉬태그 데이터 불러오기
likes = cookies["like"]
hashtags = cookies["hashtag"]

cookies.keys()
print(likes)
print(hashtags)
```

```python
# 좋아요만 뽑아오기
like = []
like50 = []
for i in likes:
    try:
        like.append(int(i[-4:-1]))
        like50.append(int(i[-4:-1]))
    except ValueError:
        like.append(0)
        like50.append(0)
    
like50.sort(reverse=True)
like50 = like50[:50]
print(like)
print(like50)
```

```python
# 해당 좋아요의 인덱스
index_l = []
for i in like50:
    index = like.index(i)
    index_l.append(index)

print(index_l)
```

```python
hashtag50 = []
for data1 in index_l:
    hashtag_l = hashtags.pop(data1)
    for data2 in hashtag_l:
        hashtag50.append(data2)
    hashtag50[:50]

print(hashtag50)
```

```python
#해시태그 개수 카운트
count = Counter(hashtag50)
#상위 50개 출력
tag_list = count.most_common(50)
print(tag_list)
```

```python
import matplotlib.pyplot as plt
#워드클라우드 만들기
#폰트, 배경, 크기 등 설정
wc = WordCloud(font_path='c:/Windows/Fonts/경기천년제목_Light.ttf',\
              background_color = 'white',\
              width = 1000,\
              height = 1000,\
              max_words = 50,\
              max_font_size = 300)
plt.figure(figsize=(15,15)) #그래프 사이즈 지정
wc.generate_from_frequencies(dict(tag_list)) #빈도수에 따라 폰트 크기결정
# wc.to_file('keyword.png') #png파일로 저장
plt.imshow(wc)
plt.axis('off') #가로,세로 축 제거
plt.show()
```
