```python
import json
from collections import Counter
from wordcloud import WordCloud
import re
```

```python
#Load in cookie
cookies = json.load(open('instagram_data_hwa_2-4gate.json', 'rb'))
# 좋아요, 해쉬태그 데이터 불러오기
likes = cookies["like"]
hashtags = cookies["hashtag"]

cookies.keys()
print(likes)
print(hashtags)
```

```python
# 좋아요만 뽑아오기
like = []
like50 = []
for i in likes:
    numbers = re.sub(r'[^0-9]',"", i)
    if numbers =="":
        like.append(0)
        like50.append(0)
    else:
        like.append(int(numbers))
        like50.append(int(numbers))

print(like)
print(like50)
like50.sort(reverse=True)
like50 = like50[:50]
```

```python
# 해당 좋아요의 인덱스
index_l = []
for i in like50:
    index = like.index(i)
    index_l.append(index)

print(index_l)
```

```python
tag_list = [] # 전체 해시태그
for i in index_l: # 해당 인덱스의
    hashtag_l = hashtags.pop(i) #해시태그를
    for j in hashtag_l: # 리스트에 하나씩 추가
        if j.count("#") >= 2:
            tag = j.split("#")
            for k in tag:
                if k != '':
                    tag_list.append(k)
        else:
            tag = j.strip("#")
            if tag != "":
                tag_list.append(tag)
```

```python
#해시태그 개수 카운트
count = Counter(tag_list)
#상위 50개 출력
hashtag50 = count.most_common(50)
```

```python
import matplotlib.pyplot as plt
#워드클라우드 만들기
#폰트, 배경, 크기 등 설정
wc = WordCloud(font_path='c:/Windows/Fonts/경기천년제목_Light.ttf',\
              background_color = 'white',\
              width = 1000,\
              height = 1000,\
              max_words = 50,\
              max_font_size = 300)
plt.figure(figsize=(15,15)) #그래프 사이즈 지정
wc.generate_from_frequencies(dict(hashtag50)) #빈도수에 따라 폰트 크기결정
# wc.to_file('keyword.png') #png파일로 저장
plt.imshow(wc)
plt.axis('off') #가로,세로 축 제거
plt.show()
```
