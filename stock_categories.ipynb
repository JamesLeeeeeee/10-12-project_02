# 다음 전종목 시세에서 업종별 종목 크롤링

```python
from selenium import webdriver
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import os
import csv
import re
import openpyxl

path = "C:\python_temp\chromedriver_win32\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("https://finance.daum.net/domestic/all_quotes")

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

html = driver.page_source
soup = BeautifulSoup(html, "html.parser")

def category(num):
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    titles = soup.find("ul", attrs={'id':'boxMarkets'}).get_text().strip().split()
    driver.find_element_by_link_text(titles[num]).click()
    time.sleep(5)
    html = driver.page_source
    soup = BeautifulSoup(html, "html.parser")
    category = soup.find_all("div", "titB")
    #업종별 이름 뽑기
    categories = []
    for i in category:
        name = re.split('\n',f'{i.get_text()}')
        name = name[1].replace('▲','').replace('▼','').replace('/',',')
        categories.append(name)

    #저장명 생성
    now = time.localtime()
    f_name = "%04d-%02d-%02d" %(now.tm_year, now.tm_mon, now.tm_mday)
    save_name = f_name+'-'+titles[num]+'.xlsx'

    writer = pd.ExcelWriter(save_name, engine='xlsxwriter')
    #pd.ExcelWriter사용 저장방법
    tables = soup.find_all("table")
    no = 0
    for i in range(len(tables)):
        if i%2 == 0:
            table = tables[i]
            table_html = str(table)
            dfs = pd.read_html(table_html)
            df = pd.DataFrame(dfs[0])
            print(f'{categories[no]}')
            print(df)
            df['종목명'].to_excel(writer, index=False, sheet_name=f'{categories[no]}')
            no += 1
        else:
             pass
    writer.save()
    time.sleep(2)
category(0) #코스피 종목
category(1) #코스닥 종목
```
