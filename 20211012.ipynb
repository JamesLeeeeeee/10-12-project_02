```python
def findx(driver, xpath):
    return driver.find_element_by_xpath(xpath)

def findsx(driver, xpath):
    return driver.find_elements_by_xpath(xpath)
```

```python
class Post: #게시물 클래스
    def __init__(self,driver):
        try:
            #필수 속성들 계정,내용, 시간, 대표이미지
            self.account = findx(driver,'/html/body/div[6]/div[2]/div/article/div/div[2]/div/div[1]/div/header/div[2]/div[1]/div[1]/span/a').text
            self.content = findx(driver,'/html/body/div[6]/div[2]/div/article/div/div[2]/div/div[2]/div[1]/ul/div/li/div/div/div[2]/span').text
            self.time = findx(driver,'/html/body/div[6]/div[2]/div/article/div/div[2]/div/div[2]/div[1]/ul/div/li/div/div/div[2]/div/div/time').get_attribute('datetime')
            self.first_image = driver.find_element_by_tag_name('img').get_attribute('src')
                                
        except: 
            print('게시글을 열어논상태에서만 가능합니다.')
            
        #없을수도 있는 속성들 해시태그, 좋아요
        try: 
            self.hash_tags = []
            self.ele_hash_tags = findsx(driver, '/html/body/div[6]/div[2]/div/article/div/div[2]/div/div[2]/div[1]/ul/div/li/div/div/div[2]/span/a')
            for ele in self.ele_hash_tags:
                self.hash_tags.append(ele.text)
            
        except:
            self.hash_tags = '해시태그 정보가 없습니다.'

        try:
            self.like = findx(driver, '/html/body/div[6]/div[2]/div/article/div/div[2]/div/div[2]/section[2]/div/div/a/span').text
        except:
            self.like = '좋아요 정보가 없습니다.'   

    def get_all_attribute(self):
        return {'계정':self.account, '내용':self.content, '게시날짜':self.time, '이미지주소':self.first_image, '해시태그': self.hash_tags, '좋아요개수':self.like}
    
    def print_all_attribute(self):
        print(f"계정 : {self.account}\n")
        print(f"내용 : {self.content}\n")
        print(f"게시날짜 : {self.time}\n")
        print(f"이미지주소 : {self.first_image}\n")
        print(f"해시태그 : {self.hash_tags}\n")
        print(f"좋아요개수 : {self.like}\n")
```

```python
#인기게시글 Post 리스트 얻기
def get_top_posts(driver):
    posts = []
    div_posts = findsx(driver,'//*[@id="react-root"]/section/main/article/div[1]/div/div/div/div/a')
    
    for div_post in div_posts:
        div_post.click()  #게시글 열기
        time.sleep(1)
        
        posts.append(Post(driver))
        
        findx(driver,'/html/body/div[6]/div[3]/button').click() #게시글 닫기
        time.sleep(1)
    
    return posts
```

```python
top_posts = get_top_posts(driver)
```

```python
i=1
for post in top_posts:
    print('인기게시글번호:',i)
    post.print_all_attribute()
    print('-'*130+'\n')
    i+=1
```

```python
#최근사진 포스트 얻기
#인기게시글 Post 리스트 얻기
def get_recent_posts(driver):
    posts = []
    div_posts = findsx(driver,'//*[@id="react-root"]/section/main/article/div[2]/div/div/div/a')
    
    for div_post in div_posts:
        div_post.click()  #게시글 열기
        time.sleep(1)
        
        posts.append(Post(driver))
        
        findx(driver,'/html/body/div[6]/div[3]/button').click() #게시글 닫기
        time.sleep(1)
    
    return posts
```
