```python
from bs4 import BeautifulSoup as bs
from selenium import webdriver
import pandas as pd

path="c:\python_temp\chromedriver.exe"
driver=webdriver.Chrome(path)
driver.get("https://finance.daum.net/")
driver.maximize_window()
```

```python
html=driver.page_source
soup=bs(html,'html.parser')
ranks=soup.find_all('div',class_="halfB")
def section_tap():
    kotitle=[]
    konum=[1,2,3,4,5]
    kosection=[]
    koupdown=[]
    koname=[]
    for i in ranks[1:3]:
        kotitle.append(i.select('h4')[0].text)
        section=(i.select('a'))[1::2]
        for j in section:
            kosection.append(j.text)
        name=(i.select('a'))[2::2]
        for k in name:
            koname.append(k.text.replace('\n',"").replace(" ","").replace("▲",""))
        updown=(i.select('span'))
        for l in updown:
            koupdown.append(l.text)
    print(kotitle)
    print(kosection)
    print(koname)
    print(koupdown)
    
    list=[]
    for i in range(10):
        list.append([kosection[i],koname[i],koupdown[i]])
    kospi=pd.DataFrame(list[0:5], columns=['업종','종목','등락'], index=konum)
    kosdak=pd.DataFrame(list[5:], columns=['업종','종목','등락'], index=konum)
    print()
    print(kotitle[0])
    print(kospi)
    print()
    print(kotitle[1])
    print(kosdak)
def rank():
    section_rank={'업종코스피':kospi,'업종코스닥':kosdak}
    what=input("")
    return section_rank[what] 
section_tap()

rank()
```
