```python
#вХёВёЮВЌљ вХѕьЋёВџћьЋю вЇ░ВЮ┤ьё░ ВДђВаЋ
stopwords=['ВЮё','віћ','ВЮ┤Ж░ђ','Ж╗ўВёю','Ж▓Ѓ','ВЮў','Ж░ђ','ВЮ┤','ВЮђ','вЊц','віћ','Вбђ','Въў','Ж▒Ї','Ж│╝','вЈё','вЦ╝','Вю╝вАю','Въљ','ВЌљ','ВЎђ','ьЋю','ьЋўвІц','вгИВЮў','ВаёьЎћ','ВБ╝вгИ','ВўѕВЋй','ьћёвАюьЋё','В╣┤ьєА','ьЋўвІе','вДЂьЂг','ьЂ┤вдГ','В▒ёвёљ','вїЊЖИђ','в░Ј']
```

```python
#ьўЋьЃюВєї вХёВёЮВЮё ьЋю ВъљвБї ВаЋВаюЖ│╝ВаЋ
train_input=[]
for sentence in df['text']:
    train_input.append([word for word in mecab.morphs(sentence)if word not in stopwords])
```

```python
train_input[:10]
```

```python
#ьєаьЂ░ьЎћ
from tensorflow.keras.preprocessing.text import Tokenizer #ьєаьЂ░ьЎћ
from tensorflow.keras.preprocessing.sequence import pad_sequences #ьїевћЕ
tokenizer= Tokenizer()
tokenizer.fit_on_texts(train_input)
print(tokenizer.word_index)
```

```python
threshold=3
words_cnt=len(tokenizer.word_index)
rare_cnt=0
words_freq=0
rare_freq=0

for key, value in tokenizer.word_counts.items():
    
    words_freq=words_freq+ value

    if value< threshold:
        rare_cnt +=1
        rare_freq= rare_freq+ value

print('ВаёВ▓┤ вІеВќ┤ Вѕў :{}'.format(words_cnt))
print(f'в╣ѕвЈёЖ░ђ {threshold-1} ВЮ┤ьЋўВЮИ вІеВќ┤ Вѕў : {rare_cnt}')
print('ьЮгЖиђ вІеВќ┤ в╣ёВюе: {}%'.format(round((rare_cnt/words_cnt)*100)))
print('ьЮгЖиђ вІеВќ┤ вЊ▒ВъЦ в╣ѕвЈё в╣ёВюе: {}'.format(round((rare_freq/words_freq)*100))) 
```

```python
vocab_size= words_cnt- rare_cnt +2
print(vocab_size)
```

```python
tokenizer=Tokenizer(vocab_size, oov_token='OOV') #ьєаьЂ░ВЌљ ВЌєвіћ вІеВќ┤віћ oov(out of vocabulary)
tokenizer.fit_on_texts(train_input)
train_input= tokenizer.texts_to_sequences(train_input)
```

```python
#ВъљвБї в░░ВЌ┤ьЎћ
train_target=np.array(df['target'])
print(train_target.shape)
print(len(train_input))
```

```python
#ьєаьЂ░ьЎћ вљю ВъљвБїЖ░ђ ВЌєвіћ ьќЅ ВаюЖ▒░
drop_train=[index for index, sentence in enumerate(train_input) if len(sentence)<1]

train_input=np.delete(train_input, drop_train, axis=0)
train_target=np.delete(train_target,drop_train, axis=0)
print(len(train_input))
print(len(train_target))
```

```python
#ВъљвБїЖИИВЮ┤ ВхювїЊЖ░њ ВёцВаЋ
max_len=1000
```

```python
#ьїевћЕВЮё ВБ╝Вќ┤ ВхювїЊЖ░њВЌљ вДъЖ▓ї вДъВХц.
train_input = pad_sequences(train_input, maxlen=max_len)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_input, test_input, train_target, test_target= train_test_split(
train_input, train_target, test_size=0.1)
```

```python
train_input, val_input, train_target, val_target=train_test_split(
train_input, train_target, test_size=0.2)
```

```python
#ваѕВЮ┤Вќ┤вЦ╝ ВБ╝Вќ┤ ьЋЎВіхьъѕЖИ░ьѓц
from tensorflow.keras.layers import Embedding, Dense, LSTM
from tensorflow.keras.models import Sequential
#ВхюВаЂВЮў вфевЇИВЮё В░ЙЖИ░ ВюёьЋю вфевЊѕ
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
model= Sequential()
model.add(Embedding(vocab_size, 100))
model.add(LSTM(128))
model.add(Dense(3,activation='softmax'))
es=EarlyStopping(monitor='val_loss', mode='min',verbose=1, patience=4)
mc=ModelCheckpoint('best_model.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True)
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy', metrics=['acc'])
history= model.fit(train_input,train_target, epochs=20, callbacks=[es, mc], batch_size=20, validation_split=0.2)
```

```python

model.evaluate(val_input, val_target)
```

```python
model.evaluate(test_input, test_target)
```

```python
def predict_(new_sentence):
    new_token= [word for word in mecab.morphs(new_sentence) if word not in stopwords]
    new_sequences= tokenizer.texts_to_sequences([new_token])
    new_pad= pad_sequences(new_sequences, maxlen=max_len)
    score= (model.predict(new_pad))
    for i in score:
        print(np.argmax(i))
    print(score)
```
