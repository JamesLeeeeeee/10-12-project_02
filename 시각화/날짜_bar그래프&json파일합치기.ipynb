```python
import json
#Load in cookie
cookies = json.load(open('instagram_data_hwa_2gate.json', 'rb'))
#날짜 데이터 불러오기
dates = cookies
```

```python
print(dates)
```

```python
#작업이 얼마나 진행됐는지 진행률프로세스바 라이브러리
from tqdm import tqdm
#년,월만 뽑아내기
date_YM = []
for i in tqdm(range(len(dates))):
    date_YM.append(dates[i][:7])
print(date_YM)
```

```python
#년만 뽑아내기
date_Y = []
for i in tqdm(range(len(dates))):
    date_Y.append(dates[i][:4])
print(date_Y)
```

```python
#월만 뽑아내기
date_M = []
for i in tqdm(range(len(dates))):
    date_M.append(dates[i][5:7])
print(date_M)
```

```python
from collections import Counter
#년,월 카운팅
count_date_YM = Counter(date_YM)
print(count_date_YM)
```

```python
#년 카운팅
count_date_Y= Counter(date_Y)
print(count_date_Y)
```

```python
#월 카운팅
count_date_M = Counter(date_M)
print(count_date_M)
```

```python
#한글 폰트 설정
import matplotlib
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()
matplotlib.rc('font', family=font_name)
```

```python
import matplotlib.pyplot as plt
#년,월 게시물 빈도수 그래프 그리기
#그래프 사이즈 지정
plt.figure(figsize=(15, 6))
#x,y축 네이밍
plt.xlabel('게시날짜')
plt.ylabel('빈도수')
#그래프에 그리드 그리기
# plt.grid(True)
#사전의 값 정렬
Sorted_Dict_Values_YM = sorted(count_date_YM.values(), reverse=True)
#사전의 키 정렬
Sorted_Dict_Keys_YM = sorted(count_date_YM, key=count_date_YM.get, reverse=True)

#bar그래프 그리기
plt.bar(range(len(count_date_YM)), Sorted_Dict_Values_YM, align='center')
plt.xticks(range(len(count_date_YM)), list(Sorted_Dict_Keys_YM), rotation='70')

plt.show()
```

```python
#년 게시물 빈도수 그래프 그리기
#그래프 사이즈 지정
plt.figure(figsize=(10, 5))
#x,y축 네이밍
plt.xlabel('게시날짜')
plt.ylabel('빈도수')
#그래프에 그리드 그리기
# plt.grid(True)
#사전의 값 정렬
Sorted_Dict_Values_Y = sorted(count_date_Y.values(), reverse=True)
#사전의 키 정렬
Sorted_Dict_Keys_Y = sorted(count_date_Y, key=count_date_Y.get, reverse=True)

#bar그래프 그리기
plt.bar(range(len(count_date_Y)), Sorted_Dict_Values_Y, align='center')
plt.xticks(range(len(count_date_Y)), list(Sorted_Dict_Keys_Y), rotation='70')

plt.show()
```
