```python
import json
#Load in cookie
data1 = json.load(open('2.JSON자료\혜화역1번출구_.json', 'rb'))
data2 = json.load(open('2.JSON자료\혜화역2번출구_.json', 'rb'))
data3 = json.load(open('2.JSON자료\혜화역3번출구_.json', 'rb'))
data4 = json.load(open('2.JSON자료\혜화역4번출구_.json', 'rb'))
#날짜 데이터 불러오기
dates1 = data1['date']
dates2 = data2['date']
dates3 = data3['date']
dates4 = data4['date']
```

```python
#년만 뽑아내기
date_Y1 = []
for i in range(len(dates1)):
    date_Y1.append(dates1[i][:4])
print(date_Y1)

date_Y2 = []
for i in range(len(dates2)):
    date_Y2.append(dates2[i][:4])
print(date_Y2)

date_Y3 = []
for i in range(len(dates3)):
    date_Y3.append(dates3[i][:4])
print(date_Y3)

date_Y4 = []
for i in range(len(dates4)):
    date_Y4.append(dates4[i][:4])
print(date_Y4)
```

# 출구별 게시글 수 비교 

```python
#한글 폰트 설정
import matplotlib
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname="c:/Windows/Fonts/H2HDRM.TTF").get_name()
matplotlib.rc('font', family=font_name)

import matplotlib.pyplot as plt
len_data1 = len(data1['id'])
len_data2 = len(data2['id'])
len_data3 = len(data3['id'])
len_data4 = len(data4['id'])

plt.figure(figsize=(6, 6))
data_len_list = [len_data1,len_data2,len_data3,len_data4]
labels = ['gate1', 'gate2', 'gate3', 'gate4']
explode = [0.05, 0.05, 0.05, 0.15]
# colors = ['plum', 'violet', 'mediumorchid', 'blueviolet']
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
patches, texts, autotexts = plt.pie(data_len_list, labels=labels, autopct='%.1f%%', startangle=90, counterclock=False,
                   explode=explode,shadow=True, colors=colors, pctdistance=0.7)

#add circle 도넛모양 만들기(중앙에 흰색으로 원 넣기)
centre_circle = plt.Circle((0,0),0.30,color='white')
plt.gca().add_artist(centre_circle)
for t in texts:
    t.set_color("dimgrey")
    t.set_fontsize(15)
for t in autotexts:
    t.set_color('white')
    t.set_fontsize(18)
    t.set_fontweight(200)
plt.savefig("출구별 활성도.png", bbox_inches="tight")
plt.show()
```
