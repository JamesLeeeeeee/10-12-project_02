```python
import json
#Load in cookie
cookies = json.load(open('instagram_data_hwa_2gate.json', 'rb'))
#해시태그 데이터 불러오기
dates = sum(cookies['hashtag'],[])
# print(dates)
tags = []
for i in dates:
    tag = i.split('#')
    for j in tag:
        if '' == j:
            pass
        else:
            tags.append(j)
print(tags)
```

```python
from collections import Counter
from wordcloud import WordCloud
```

```python
#해시태그 개수 카운트
count = Counter(tags)
#상위 50개 출력
tag_list = count.most_common(50)
print(tag_list)
```

```python
import matplotlib.pyplot as plt
#워드클라우드 만들기
#폰트, 배경, 크기 등 설정
wc = WordCloud(font_path='c:/Windows/Fonts/malgun.ttf',\
              background_color = 'white',\
              width = 1000,\
              height = 1000,\
              max_words = 50,\
              max_font_size = 300)
plt.figure(figsize=(15,15)) #그래프 사이즈 지정
wc.generate_from_frequencies(dict(tag_list)) #빈도수에 따라 폰트 크기결정
# wc.to_file('keyword.png') #png파일로 저장
plt.imshow(wc)
plt.axis('off') #가로,세로 축 제거
plt.show()
```

# 이미지 맵핑, 이미지색상으로 워드클라우드만들기

```python
import numpy as np # 이미지 데이터를 다루기 위해
import pandas as pd # 데이터를 다루기 위한 라이브러리
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator # wordcloud 라이브러리
from PIL import Image # 이미지를 위한 라이브러리
import matplotlib.pyplot as plt # 그래프 및 이미지화 라이브러리
```

```python
stopwords = set(STOPWORDS)
stopwords.add('아이돌도시락') # 제거할 단어들.. 적용이 안되네..ㅠ
wine_mask = np.array(Image.open('hwa_map.jpg'))
wordcloud = WordCloud(font_path='c:/Windows/Fonts/malgun.ttf', background_color='white',
                      stopwords=stopwords,
                      mask=wine_mask,
                      width = 1000,
                      height = 1000,
                      max_words = 50,
                      max_font_size = 300).generate_from_frequencies(dict(tag_list))
image_colors = ImageColorGenerator(wine_mask)
wordcloud = wordcloud.recolor(color_func=image_colors)
plt.figure(figsize=(15,15))
plt.axis('off')
plt.imshow(wordcloud, interpolation='bilinear') 
plt.show()
```
