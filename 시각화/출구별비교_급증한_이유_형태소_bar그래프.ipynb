# 출구별 게시글현황에서 2018년도에 급증한 거 확인

```python
import json
#Load in cookie
cookies = json.load(open('2.JSON자료\instagram_data_hwa_1gate.json', 'rb'))
#리스트에 담긴 text 하나로 합치기
dates = cookies['date']
```

```python
index_2019=[]
index_2018=[]
index_2017=[]
for i in range(len(dates)):
    if '2019' in dates[i]:
        index_2019.append(i)
    if '2018' in dates[i]:
        index_2018.append(i)
    if '2017' in dates[i]:
        index_2017.append(i)
```

```python
print(index_2019)
```

```python
texts = cookies['text']
text_list_2019 = []
text_list_2018 = []
text_list_2017 = []
for i in index_2019:
    text_list_2019.append(texts[i])
for i in index_2018:
    text_list_2018.append(texts[i])
for i in index_2017:
    text_list_2017.append(texts[i])    
text_list_2019 = sum(text_list_2019,[])
text_list_2018 = sum(text_list_2018,[])
text_list_2017 = sum(text_list_2017,[])
```

```python
print(text_list_2019)
```

```python
import re
#정규식으로 불필요한 문자 제거
newtexts_2019 = ''
for text in text_list_2019:
    text = text = re.sub('[0-9]+', '', str(text))
    text = re.sub('[A-Za-z]+', '', text)
    newtext = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ·!』\\‘’|\(\)\[\]\<\>`\'…》]', '', text)
    newtexts_2019 += newtext

newtexts_2018 = ''
for text in text_list_2018:
    text = text = re.sub('[0-9]+', '', text)
    text = re.sub('[A-Za-z]+', '', text)
    newtext = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ·!』\\‘’|\(\)\[\]\<\>`\'…》]', '', text)
    newtexts_2018 += newtext

newtexts_2017 = ''
for text in text_list_2017:
    text = text = re.sub('[0-9]+', '', text)
    text = re.sub('[A-Za-z]+', '', text)
    newtext = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ·!』\\‘’|\(\)\[\]\<\>`\'…》]', '', text)
    newtexts_2017 += newtext
```

```python
print(newtexts_2019)
```

```python
import matplotlib.pyplot as plt
from konlpy.tag import Okt
okt = Okt()
nouns_list_2019 = okt.nouns(newtexts_2019) #명사추출
nouns_list_2018 = okt.nouns(newtexts_2018) #명사추출
nouns_list_2017 = okt.nouns(newtexts_2017) #명사추출
```

```python
from collections import Counter
#단어 개수 세기
count_2019 = Counter(nouns_list_2019)
count_2018 = Counter(nouns_list_2018)
count_2017 = Counter(nouns_list_2017)
```

```python
print(count_2019)
```

```python
#count로 빈도수 카운트한 단어 for문 돌려서 사전형태로 만들기
words_2019 = dict()
#명사 빈도가 높은 n개 추출.most_common(n)
noun_list_2019 = count_2019.most_common(30)
for nouns, counts in noun_list_2019:
    if len(nouns) > 1:
        words_2019[nouns] = counts
del(words_2019['배우'])
# del(words['위치'])
  
words_2018 = dict()
noun_list_2018 = count_2018.most_common(30)
for nouns, counts in noun_list_2018:
    if len(nouns) > 1:
        words_2018[nouns] = counts
        
words_2017 = dict()
noun_list_2017 = count_2017.most_common(30)
for nouns, counts in noun_list_2017:
    if len(nouns) > 1:
        words_2017[nouns] = counts
del(words_2017['문의'])
del(words_2017['인스타'])
```

```python
print('2019년 게시글 내용:',words_2019,'\n')
print('2018년 게시글 내용:',words_2018,'\n')
print('2017년 게시글 내용:',words_2017)
```
